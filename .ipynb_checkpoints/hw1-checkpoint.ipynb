```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
np.random.seed(42)
```

```python
X1 = [[1, 1]] + np.random.randn(100, 2) * 0.5
X2 = [[-1, 1]] + np.random.randn(100, 2) * 0.5
X3 = [[-1, -1]] + np.random.randn(100, 2) * 0.5

X = np.vstack([X1, X2, X3])
X = np.hstack([X, np.ones((300, 1))])
y = np.hstack([np.full(100, 0), np.full(100, 1), np.full(100, 2)])
```

```python
plt.plot(X[y==0, 0], X[y==0, 1], 'o', color='blue', label='класс 0')
plt.plot(X[y==1, 0], X[y==1, 1], 'o', color='red', label='класс 1')
plt.plot(X[y==2, 0], X[y==2, 1], 'o', color='orange', label='класс 2')

plt.legend(loc='best')
plt.show()
```

```python
models = []
model = LogisticRegression(random_state=42, max_iter=2000, fit_intercept=False)
_ = model.fit(X, np.array(y==0, dtype=np.int64))
models.append(model)

model = LogisticRegression(random_state=42, max_iter=2000, fit_intercept=False)
_ = model.fit(X, np.array(y==1, dtype=np.int64))
models.append(model)

model = LogisticRegression(random_state=42, max_iter=2000, fit_intercept=False)
_ = model.fit(X, np.array(y==2, dtype=np.int64))
models.append(model)
```

```python
w1_1, w1_2, w1_3 = models[0].coef_[0]
w2_1, w2_2, w2_3 = models[1].coef_[0]
w3_1, w3_2, w3_3 = models[2].coef_[0]
```

```python
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 3)

plt.plot(X[y==0, 0], X[y==0, 1], 'o', color='blue', label='класс 0')
plt.plot(X[y==1, 0], X[y==1, 1], 'o', color='red', label='класс 1')
plt.plot(X[y==2, 0], X[y==2, 1], 'o', color='orange', label='класс 2')

x1 = np.linspace(-10, 2)
y1 = - (w1_1 * x1 + w1_3) / w1_2
plt.plot(x1, y1, color='blue')

x2 = np.linspace(-10, 2)
y2 = - (w2_1 * x2 + w2_3) / w2_2
plt.plot(x2, y2, color='red')

x3 = np.linspace(-10, 10)
y3 = - (w3_1 * x3 + w3_3) / w3_2
plt.plot(x3, y3, color='orange')

plt.legend(loc='best')
plt.show()
```
